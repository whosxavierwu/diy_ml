```python
input_dim = (128, 128, 3)
batch_size = 32
# data_folder = '../../../../datasets'
data_folder = 'D:\\wuzewei\\datasets\\CelebA\\Img\\img_align_celeba\\img_align_celeba'
filenames = np.array(glob(os.path.join(data_folder, '*.jpg')))
# os.listdir(data_folder)
print(len(filenames))
print(filenames[:3])
```

```python
def peek_some_pictures(filenames, num):
    tmp_filenames = np.random.choice(filenames, num)
    figure = plt.figure(figsize=(3, 3))
    nrows = int(np.ceil(np.sqrt(num)))
    ncols = nrows
    plt.axis("off")
    for idx, tmp_filename in enumerate(tmp_filenames):
        im_lol1 =  plt.imread(tmp_filename)
        ax = figure.add_subplot(nrows, ncols, idx+1)
        plt.axis('off')
        ax.imshow(im_lol1)
        ax.set_title('lol image 1')
    plt.show()
    return 

peek_some_pictures(filenames, 3)
# plt.show()
```
