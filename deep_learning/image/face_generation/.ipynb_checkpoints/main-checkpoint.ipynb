```python
import os
from glob import glob
import numpy as np
from matplotlib import pyplot as plt
import random
from PIL import Image

from keras.preprocessing.image import ImageDataGenerator
from Autoencoder import Autoencoder
```

```python

data_folder = 'D:\\wuzewei\\datasets\\CelebA\\Img\\img_align_celeba\\img_align_celeba'
filenames = np.array(glob(os.path.join(data_folder, '*.jpg')))
print(len(filenames))
print(filenames[:3])
```

```python
INPUT_DIM = (128, 128, 3)
Z_DIM = 200
BATCH_SIZE = 128
NUM_EPOCHS = 10
```

```python

vae = Autoencoder(
        input_dim=INPUT_DIM,
        z_dim=Z_DIM,
        encoder_conv_settings=[
            [32, 3, 2],
            [64, 3, 2],
            [64, 3, 2],
            [64, 3, 2],
        ],
        decoder_conv_settings=[
            [64, 3, 2],
            [64, 3, 2],
            [32, 3, 2],
            [3,  3, 2],
        ],
        is_variational=True, 
        use_batch_norm=True,
        use_dropout=True
    )
# print(vae.encoder.summary())
# print(vae.decoder.summary())
vae.compile(0.0005, 10000)
```

```python
data_gen = ImageDataGenerator(
    rescale=1./255
)
data_flow = data_gen.flow_from_directory(
    'D:\\wuzewei\\datasets\\CelebA\\Img\\img_align_celeba',
    target_size=INPUT_DIM[:-1],
    batch_size=BATCH_SIZE,
    shuffle=True,
    class_mode='input',
    subset='training'
)
```

```python
vae.train_with_generator(     
    data_flow,
    epochs = NUM_EPOCHS,
    steps_per_epoch = len(filenames) / BATCH_SIZE,
    run_folder = 'tmp/',
    print_every_n_batches = 100,
    initial_epoch = 0
)
```
